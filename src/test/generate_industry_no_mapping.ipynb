```python
def generate_industry_no_mapping(tree, d={}):
    ''' According to a tree, generate "industry_no" mapping dictionary.
    '''
    d = deepcopy(d)
    if 'no' in tree.keys():
        d[tree['no']] = tree.get('des', '')
        
    if 'n' in tree.keys():
        for node in tree['n']:
            d[node['no']] = node.get('des', '')
            if 'n' in node.keys():
                d = generate_industry_no_mapping(node, d)
    return d
```
