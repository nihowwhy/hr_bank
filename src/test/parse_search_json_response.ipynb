```python
def get_flat_dict(d, return_dict={}):
    for key, value in d.items():
        if isinstance(value, dict):
            return_dict = get_flat_dict(value, return_dict)
        else:
            return_dict[key] = value
    return return_dict
            
for k, v in get_flat_dict(j).items():
    print(k)
    print(v)
    print()
```
